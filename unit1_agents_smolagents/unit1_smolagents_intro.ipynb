## What is smolagents?

`smolagents` is a lightweight agent library by Hugging Face that makes it easy to build agents with:
- Any LLM (HF Inference API, OpenAI, Anthropic, local models)
- Custom tools
- Code execution capabilities

Two main agent types:
- **CodeAgent**: Writes and executes Python code as actions
- **ToolCallingAgent**: Uses JSON tool calls (like OpenAI function calling)

## The Agent Loop

The core of `smolagents` is the ReAct (Reasoning + Acting) loop:

```
1. Receive task
2. Think: what tool should I call?
3. Act: call the tool
4. Observe: see the result
5. Repeat until done
```

````python
from smolagents import tool

# Define a simple custom tool
@tool
def add(a: int, b: int) -> int:
    """Adds two integers together.
    
    Args:
        a: First integer
        b: Second integer
    """
    return a + b

print("Tool 'add' created successfully")
print(f"Test: add(3, 4) = {add(3, 4)}")
````

````python
# Inspect the tool schema (what the LLM sees)
print(f"Tool name: {add.name}")
print(f"Tool description: {add.description}")
print(f"Tool inputs: {add.inputs}")
````

## Building a CodeAgent

A `CodeAgent` writes Python code to solve tasks instead of calling JSON tool schemas. This is more flexible and powerful.

````python
from smolagents import CodeAgent, HfApiModel, DuckDuckGoSearchTool

# Initialize the LLM
model = HfApiModel(model_id="Qwen/Qwen2.5-Coder-32B-Instruct")

# Create the agent with tools
agent = CodeAgent(
    tools=[DuckDuckGoSearchTool(), add],
    model=model
)

print(f"Agent created with HfApiModel (Qwen/Qwen2.5-Coder-32B-Instruct)")
print(f"Tools available: DuckDuckGoSearchTool, add")
````

## Key Concepts: Tool Calling vs Code Agents

| | ToolCallingAgent | CodeAgent |
|---|---|---|
| Action format | JSON schema | Python code |
| Flexibility | Medium | High |
| Safety | Higher (predefined tools) | Requires sandbox |
| Best for | Structured tool use | Complex multi-step logic |

My Space uses `ToolCallingAgent` since it integrates well with Gradio and has predictable tool usage.
