## Paso 1 — Entender la estructura del template

El template `agents-course/First_agent_template` tiene esta estructura:

```
First_agent_template/
├── app.py           ← Punto de entrada: configura el agente y lanza Gradio
├── Gradio_UI.py     ← Interfaz de usuario con Gradio
├── agent.json       ← Configuración del agente (herramientas, modelo)
├── prompts.yaml     ← System prompts del agente
├── requirements.txt ← Dependencias
└── tools/           ← Herramientas personalizadas
```

## Paso 2 — Configuración del agente en `app.py`

El archivo clave es `app.py`. Aquí se configura el modelo, las herramientas y se lanza la UI.

````python
# Estructura del app.py del Space desplegado
app_py_estructura = """
from smolagents import ToolCallingAgent, HfApiModel, DuckDuckGoSearchTool
from smolagents import load_tool
from Gradio_UI import GradioUI

# 1. Definir el modelo LLM a usar
model = HfApiModel(model_id="Qwen/Qwen2.5-72B-Instruct")

# 2. Cargar herramientas
image_generation_tool = load_tool("m-ric/text-to-image", trust_remote_code=True)

# 3. Crear el agente con las herramientas
agent = ToolCallingAgent(
    tools=[DuckDuckGoSearchTool(), image_generation_tool],
    model=model
)

# 4. Lanzar la UI con Gradio
GradioUI(agent).launch()
"""

print("Estructura de app.py explicada")
print(app_py_estructura)
````

## Paso 3 — El ciclo Pensamiento → Acción → Observación con smolagents

Con `smolagents`, el ciclo del agente es automático. El agente:
1. Recibe la pregunta del usuario
2. **Piensa** qué herramienta usar
3. **Actúa** llamando a la herramienta
4. **Observa** el resultado
5. Repite hasta tener la respuesta final

````python
print("Diferencia entre ToolCallingAgent y CodeAgent:")
print()
print("ToolCallingAgent:")
print("  - Usa JSON para llamar herramientas (estilo OpenAI function calling)")
print("  - Más predecible y seguro")
print("  - Ideal para integraciones con Gradio")
print()
print("CodeAgent:")
print("  - Escribe código Python como acción")
print("  - Más flexible y potente")
print("  - Requiere entorno de ejecución seguro (sandbox)")
````
