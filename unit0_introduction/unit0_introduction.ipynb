## What is an AI Agent?

An **AI Agent** is a system that uses an LLM as its reasoning engine and takes actions in the world through **tools**.

Key components:
- **Brain (LLM):** The language model that decides what to do
- **Tools:** Functions the agent can call (web search, calculator, code executor, etc.)
- **Memory:** Short-term (context window) and long-term (vector stores)
- **Actions:** What the agent executes based on its reasoning

The core **agent loop**:
```
Observe → Think → Act → Observe → Think → Act → ...
```

## LLM vs Agent

| Feature | LLM | Agent |
|---|---|---|
| Input | Text prompt | Observations from environment |
| Output | Text | Actions (tool calls, text) |
| Memory | Context window only | Can have persistent memory |
| Can act? | No | Yes, through tools |
| Multi-step? | No | Yes |

## Types of Agent Memory

1. **In-context storage** – Information within the active context window
2. **External storage** – Databases outside the model (fuzzy via embeddings, or exact via key-value)
3. **In-weights storage** – Knowledge baked into model parameters during training
4. **In-cache storage** – Saved computation states (KV cache)

## Types of Agent Actions

Agents can take different types of actions:
- **Storage manipulation** – Read/write files, databases, code
- **Process execution** – Terminal commands, test runners
- **UI interaction** – Web browsers, desktop GUIs
- **Service calls** – APIs, IoT sensors, data streams
- **Cross-agent interactions** – Spawning subagents, calling other agents
